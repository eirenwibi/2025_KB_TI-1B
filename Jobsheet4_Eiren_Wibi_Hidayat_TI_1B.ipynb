#**3. Membuat DataFrame**

##a. Membuat DataFrame

```python
import pandas as pd

df = pd.DataFrame(data= {"Nama" : ["Ali","Joko","Adi"],
                         "Umur" : [12, 13, 15],
                         "Kelas" : [6, 7, 8]})

df.columns
df.columns.values[0] = "Nama Singkat"
df

data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

df = pd.DataFrame(data)
df
```

##b. Mengakses Elemen pada DataFrame

```python
df.iloc[0:2, [0,1]]
df.iloc[0:2, 0:2]
df.iloc[:, 0:2]
df.loc[0:2, 'calories']
df.loc[0:2, ['calories', 'duration']]
df['calories']
df['calories'][0]
df[['calories']]
df[['calories']].loc[[0, 1]]
```

##c. Memberi Nama pada Index

```python
data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ['day1', 'day2', 'day3'])
print(df)
df
```
