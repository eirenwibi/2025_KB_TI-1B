## Mengatasi Missing Values dengan Drop Data

### 1. Hapus Data yang kosong (NaN)

```python
df = pd.read_csv('./Data.csv')
df.dropna(inplace=True)

df
```

## Mengatasi Missing Values dengan Nilai Baru

### 2. Ganti Data yang kosong (NaN) menjadi : 

### <ul><li>Umur = 50 </li><li>Gaji = 25000000</li></ul>

```python
df = pd.read_csv('./Data.csv')
df['Umur'].fillna(50, inplace=True)
df['Gaji'].fillna(25000000, inplace=True)
df
```

## Mengatasi Missing Values dengan Nilai Mean

## 3. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Rata-Rata

```python
df = pd.read_csv('./Data.csv')
mean_umur = df['Umur'].mean();
mean_gaji = df['Gaji'].mean();

df['Umur'].fillna(mean_umur, inplace=True)
df['Gaji'].fillna(mean_gaji, inplace=True)
```

```python
df
```

## Mengatasi Missing Values dengan Nilai Median

## 4. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Median

```python
df = pd.read_csv('./Data.csv')
mean_umur = df['Umur'].median();
mean_gaji = df['Gaji'].median();

df['Umur'].fillna(mean_umur, inplace=True)
df['Gaji'].fillna(mean_gaji, inplace=True)

df
```

## Mengatasi Missing Values dengan Nilai Mode

## 5. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Modus

```python
df = pd.read_csv('./Data.csv')
mean_umur = df['Umur'].mode()[0];
mean_gaji = df['Gaji'].mode()[0];

df['Umur'].fillna(mean_umur, inplace=True)
df['Gaji'].fillna(mean_gaji, inplace=True)

df
```

## Mengatasi Wrong Data dengan Ubah Data

## 6. Apabila Ada Umur yang diatas 120 maka ubah menjadi 33

```python
df = pd.read_csv('./Data.csv')

for x in df.index:
    if df.loc[x, 'Umur'] > 120:
        df.loc[x, 'Umur'] = 33
        
df
```

## Soal-2
Buatlah list Top 5 Gas stations yang memiliki nilai transaksi paling banyak
---

```python
import pandas as pd

# Load data
transactions = pd.read_csv('sample_data/transactions_1k.csv')
gasstations = pd.read_csv('sample_data/gasstations.csv')

# Merge data
merged_gas = pd.merge(transactions, gasstations, on='GasStationID')

# Hitung total transaksi per gas station
top_gasstations = merged_gas.groupby('GasStationID').agg(
    Amount=('Amount', 'sum'),  # Total liter/minyak terjual
    Gas_Sales=('Price', 'sum')  # Total nilai transaksi
).sort_values('Gas_Sales', ascending=False).head(5)

# Format output sesuai gambar
top_gasstations = top_gasstations.rename(columns={'Gas_Sales': 'Gas Sales'})
top_gasstations.index.name = 'Gas Stations'

# Tampilkan hasil
top_gasstations
```

## Soal-3
Buatlah list Top 5 Jenis produk yang memiliki nilai transaksi paling banyak
---

```python
import pandas as pd

# Load data
transactions = pd.read_csv('sample_data/transactions_1k.csv')
products = pd.read_csv('sample_data/products.csv')

# Merge data
merged_products = pd.merge(transactions, products, on='ProductID')

# Hitung total transaksi per produk
top_products = merged_products.groupby('Description').agg(
    Amount=('Amount', 'sum'),
    Sales=('Price', 'sum')
).sort_values('Sales', ascending=False).head(5)

# Tampilkan hasil
top_products
```

## Soal-4
Buatlah deskripsi statistik untuk masing masing hari (23,24,25,26) pada dataset transaction_1k
---

```python
import pandas as pd

# Load data
df = pd.read_csv('sample_data/transactions_1k.csv')

# Konversi kolom Date ke datetime dan ekstrak hari
df['Date'] = pd.to_datetime(df['Date'])
df['Day'] = df['Date'].dt.date  # Mengambil tanggal tanpa waktu

# Hitung statistik deskriptif untuk Amount dan Price (Sales) per hari
stats_per_day = df.groupby('Day').agg({
    'Amount': ['count', 'mean', 'std', 'min', lambda x: x.quantile(0.25), 'median', lambda x: x.quantile(0.75), 'max'],
    'Price': ['count', 'mean', 'std', 'min', lambda x: x.quantile(0.25), 'median', lambda x: x.quantile(0.75), 'max']
})

# Rename kolom untuk kejelasan
stats_per_day.columns = [
    'Amount count', 'Amount mean', 'Amount std', 'Amount min', 'Amount 25%', 'Amount 50%', 'Amount 75%', 'Amount max',
    'Sales count', 'Sales mean', 'Sales std', 'Sales min', 'Sales 25%', 'Sales 50%', 'Sales 75%', 'Sales max'
]

# Tampilkan hasil statistik
stats_per_day
```

## Soal-5
Carilah waktu terbaik (hari dan jam) dimana paling banyak user gas station melakukan transaksi
---

```python
import pandas as pd

# Buat kolom 'Gas Sales'
df_combined["Gas Sales"] = df_combined["Amount"] * df_combined["Price"]

# Ekstrak jam dari kolom 'Time'
df_combined["Hour"] = pd.to_datetime(df_combined["Time"], format="%H:%M:%S").dt.hour

# Group by Date dan Hour, lalu agregasi Amount dan Gas Sales
result = df_combined.groupby(["Date", "Hour"]).agg({
    "Amount": "sum",
    "Gas Sales": "sum"
})

# Tampilkan hasil
result
```
